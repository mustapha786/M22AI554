```
import numpy as np
import matplotlib.pyplot as plt
```

```

positive = np.array([[1,1],[0.2,0.2],[0.9,0.5]])
negative = np.array([[-1,-1],[0,0.5],[0.1,0.5]])
data= np.concatenate((positive, negative), axis=0) # data is 6 rows and 2 columns
print(data)
#data = data1 - np.mean(data1,axis=0) # will make data mean = 0
#print(data)
y=[1,1,1, -1, -1, -1]
print(y)
```

```
plt.scatter(np.array(data[0:3,0]), np.array(data[0:3,1]), marker='+', label='positive')
plt.scatter(np.array(data[3:6,0]), np.array(data[3:6,1]), marker='*', label='negative')
#y1 = y 
#plt.plot(y, y1)
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend()
plt.show()
```

```

def _Perceptron(data,w):
#print(w)
 counter = 0
 correctClassified = 0
 list1 = []
 while (correctClassified != len(data)): #Until everything is classified
  counter = counter + 1
  for sample in range(len(data)):
    x = np.append(data[sample,0:2],1)
    if y[sample]==1: #Sample is positive 
       if np.dot(np.transpose(w),x)>=0: #WX >= 0
         correctClassified=correctClassified+1
         #print("sample is pos")
       else: 
          #print("weight updated for positive missclassification:",w)
          w=w+x
          list1.append(w)
          #print("Post updated weight_pos:",w)
    else: #Sample is Negative 
         if np.dot(np.transpose(w),x)<0: # WX < 0
           correctClassified=correctClassified+1
           #print("sample is neg")
         else: 
           #print("weight updated for negative missclassification:",w)
           #print("input:",x)
           w=w-x
           #print("Post updated weight_neg:",w)
           list1.append(w)
    
  if(correctClassified != len(data)):
      correctClassified=0  
      counter = counter + 1         
 return w , list1 , counter  
```

```
def _Pred(data,wf):
  Y_Pred = []
  for sample in range(len(data)):
    x = np.append(data[sample,0:2],1)
    Y = np.dot(np.transpose(wf),x)
    Y_Pred.append(Y)
  return Y_Pred
```

```
correctClassified = 0
w0 = np.array([1,1]) # initialized w with 1
#print(w)
w0 = np.append(w0,1) # append w0=1
wf , listw,counter = _Perceptron(data , w0 )
print(wf,counter)
```
